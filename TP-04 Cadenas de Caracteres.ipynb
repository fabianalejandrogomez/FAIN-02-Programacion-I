```python
"""
Desarrollar una función que determine si una cadena de caracteres es capicúa, 
sin utilizar cadenas auxiliares. Escribir además un programa que permita verificar su funcionamiento.
"""

def capicua(cadena):
    """
    Recibe una cadena y devuelve True si es capicua
    """
    if cadena == cadena[::-1]:
        return True
    return False

if __name__ == '__main__':
    print(capicua(input("Ingrese una cadena para verificar si es capicua: ")))
```

```python
"""
Leer una cadena de caracteres e imprimirla centrada en pantalla. Suponer que la misma tiene 80 columnas.
"""


ANCHO_PANTALLA = 80


def centrar(cadena):
    espacios = (ANCHO_PANTALLA - len(cadena)) // 2
    espacios = " " * espacios
    return espacios + cadena + espacios


if __name__ == "__main__":
    print(centrar(input("Ingrese una cadena para imprirla centrada:\n")))
```

```python
"""
Los números de claves de dos cajas fuertes están intercalados dentro de un número entero llamado "clave maestra". 
Realizar un programa para obtener ambas claves, donde la primera se construye con los dígitos impares de la clave maestra 
y la segunda con los dígitos pares. Los dígitos se numeran desde la izquierda. 
Ejemplo: Si clave maestra = 12345, la clave 1 sería 135 y la  clave 2 sería 24.
"""

def get_claves(clavemaestra):
    clavemaestra = str(clavemaestra)
    return [clavemaestra[::2], clavemaestra[1::2]]


if __name__ == "__main__":
    clavemaestra = get_claves(12345)
    print("La clave 1 es " + clavemaestra[0] + " y la clave 2 es " + clavemaestra[1])
```

```python
# Practica 4 - Ejercicio 5

def filtrar_palabras1(frase, n):
    """ Filtra palabras con menos de N caracteres utilizando ciclos normales """
    # Convertimos la frase en una lista de palabras
    lista = frase.split( )
    # y concatenamos a frase aquellas palabras que tengan más de N caracteres
    frase = ""
    for palabra in lista:
        if len(palabra)>n:
            frase = frase + palabra + " "
    return frase

def filtrar_palabras2(frase, n):
    """ Filtra palabras con menos de N caracteres utilizando listas por comprensión """
     # Convertimos la frase en una lista de palabras
    lista1 = frase.split( )
    # y agregamos a lista2 aquellas palabras que tengan más de N caracteres
    # No se puede usar una cadena en lugar de una lista debido a las listas por comprensión
    lista2 = [palabra for palabra in lista1 if len(palabra)>n]
    # Por último convertimos la nueva lista en una cadena
    frase = " ".join(lista2)
    return frase

def filtrar_palabras3(frase, n):
    """ Filtra palabras con menos de N caracteres utilizando la función filter() """
    # Convertimos la frase en una lista de palabras
    listadepalabras = frase.split( )
    # y utilizamos filter() para añadir sólo las palabras que tengan más de N caracteres.
    # La lista devuelta por filter() es convertida en cadena a través del método join()
    frase = " ".join(filter(lambda x:len(x)>n, listadepalabras))
    return frase

# Programa principal
frase = input("Ingrese una frase: ")
largo = int(input("Ingrese la longitud deseada: "))
nueva = filtrar_palabras1(frase, largo)
print(nueva)
nueva = filtrar_palabras2(frase, largo)
print(nueva)
nueva = filtrar_palabras3(frase, largo)
print(nueva)
```

```python
"""
Escribir una función que reciba como parámetro una cadena de caracteres 
en la que las palabras se encuentran separadas por uno o más espacios. 
Devolver otra cadena con las palabras ordenadas alfabéticamente, 
dejando un espacio entre cada una.
"""

def ordAlfabeticamente(cadena):
    cadena = cadena.split(' ')
    cadena.sort()
    return (' ').join(cadena)

if __name__ == "__main__":
    cadena = "on any issue or pull request to go back to the pull request listing page"
    print("La oracion original:\n" + cadena, end = "\n\n")
    print("La oracion con las palabras ordenadas alfabeticamente:\n" + ordAlfabeticamente(cadena))
```

```python
"""
Desarrollar una función que devuelva una subcadena con los últimos N caracteres de una cadena dada. 
La cadena y el valor de N se pasan como parámetros.
"""

def subcadena(cadena, n):
    return cadena[len(cadena)-n:]


if __name__ == "__main__":
    print(subcadena("Hola Mundo!", 6))
```

```python
"""
Escribir un programa que permita ingresar una cadena de caracteres e imprima un mensaje indicando cuántas letras y cuántos números contiene. 
Por ejemplo, si se ingresa "Hola Mundo 123" debe indicar que se ingresaron 9 letras y 3 números
"""

import string


def Letras_y_Numeros(cadena):
    """
    Imprime por consola: 
    La cantidad de letras y números recibidos en una cadena de caracteres
    """
    # contadores
    letras, numeros = 0, 0
    for i in range(len(cadena)):
        # si es alfabetico lo cuenta
        if cadena[i].isalpha():
            letras += 1
        # si es número lo cuenta
        elif cadena[i].isdigit():
            numeros += 1
    # resultado
    print("Se ingresaron {0} letras y {1} números." .format(letras, numeros))


if __name__ == "__main__":
    # imprimo el docstring con la consigna para el usuario
    print(Letras_y_Numeros.__doc__)
    Letras_y_Numeros(input("Ingrese por favor: "))
```

```python
"""
Escribir un programa que permita ingresar una cadena de caracteres y 
coloque en mayúscula la primera letra posterior a un espacio, 
eliminando todos los espacios que contenga. 
Imprimir por pantalla la cadena obtenida.
"""


def mayusculasPorEspacio(cadena):
    nuevaCadena, continuar = '', False
    for i in range(len(cadena)):
        if continuar == True:
            continuar = False
            continue
        if cadena[i] == ' ' and i < len(cadena):
            nuevaCadena += str.upper(cadena[i+1])
            continuar = True
        else:
            nuevaCadena += cadena[i]
    return nuevaCadena


if __name__ == "__main__":
    print(mayusculasPorEspacio(input("Ingrese una cadena: ")))
```
