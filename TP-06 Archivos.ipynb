```python
def sinComentarios(programa):
    comillasAbiertas = False
    comentarioAbierto = False
    lineas = ""
    try:
        archivo = open(PATH+programa, MODO_LECTURA)
        nuevo = open(PATH_ESCRITURA+"ejercicio01.py", MODO_ESCRITURA)
        caracter = archivo.read(1)
        lineas = archivo.readlines()
        if caracter == "'" or caracter == '"' and comillasAbiertas == False:
            comillasAbiertas == True
        if caracter == "'" or caracter == '"' and comillasAbiertas == True:
            comillasAbiertas == False
        if caracter == "#" or caracter == '"' and comillasAbiertas != True:
            caracter = ""
            comentarioAbierto = True
        if comentarioAbierto == True and caracter != "\n":
            caracter = ""
        if comentarioAbierto == True and caracter == "\n":
            comentarioAbierto = False
        lineas += caracter
        nuevo.writelines(lineas)
        archivo.close()
        nuevo.close()
    except NameError as error:
        print("Error de Nombre " + error)
    except IOError as error:
        print("Error de IO " + error)
    except TypeError as error:
        print("Error de tipo de datos " + error)


if __name__ == "__main__":
    sinComentarios(input("Ingrese el nombre del archivo PY: "))
```

```python
"""
Escribir un programa que permita grabar un archivo los datos de lluvia caída durante un año.
Cada línea del archivo se grabará con el siguiente formato:
    <dia>;<mes>;<lluvia caída en mm>  por ejemplo  25;5;319
Los datos se generarán mediante números al azar, asegurándose que las fechas sean válidas.
La cantidad de registros también será un número al azar entre 50 y 200.
Por último se solicita mostrar un informe en formato matricial donde cada columna represente a un mes y cada fila corresponda a los días del mes.
Imprimir además el total de lluvia caída en todo el año.
"""


PATH = "C:\\repos\\UADE\\Programacion-I\\TP06\\input\\"
PATH_ESCRITURA = "C:\\repos\\UADE\\Programacion-I\\TP06\\output\\"
MODO_ESCRITURA = "wt"
MODO_LECTURA = "rt"


import random


def datosRandom():
    """
    Genera datos random de lluvia por día, devuelve: 
    dia - mes - mm
    """
    return random.randint(1, 31), random.randint(1, 12), random.randint(0, 1000)


def abrir(nombre):
    """
    Abre un archivo en modo escritura
    """
    try:
        return open(PATH_ESCRITURA+ nombre + ".txt", MODO_ESCRITURA)
    except IOError as error:
        print("Error al abrir el archivo: ", error)
    except NameError as error:
        print("Error en el nombre del archivo", error)


def escribirRegistro(cantidad_registros):
    archivo = abrir("registro")
    informe_mes, lluviaAnual = {}, 0
    for i in range(cantidad_registros):
        dia, mes, mm = datosRandom()
        archivo.write(str(dia) + ";" + str(mes) + ";" + str(mm) + "\n")
        informe_mes.update({mes: dia})
        lluviaAnual += mm
    print("Enero: {0}".format(informe_mes[1]))
    print("Febrero: {0}".format(informe_mes[2]))
    print("Marzo: {0}".format(informe_mes[3]))
    print("Abril: {0}".format(informe_mes[4]))
    print("Mayo: {0}".format(informe_mes[5]))
    print("Junio: {0}".format(informe_mes[6]))
    print("Julio: {0}".format(informe_mes[7]))
    print("Agosto: {0}".format(informe_mes[8]))
    print("Septiembre: {0}".format(informe_mes[9]))
    print("Octubre: {0}".format(informe_mes[10]))
    print("Noviembre: {0}".format(informe_mes[11]))
    print("Diciembre: {0}".format(informe_mes[12]))
    print("Lluvia Anual: {0}".format(lluviaAnual))
    archivo.close()



if __name__ == "__main__":
    escribirRegistro(random.randint(50, 200))
```

```python
# Practica 6 - Ejercicio 4

try:
    entrada = open("Apellidos.txt")
    armenia = open("armenia.txt", "wt")
    italia = open("italia.txt", "wt")
    españa = open("españa.txt", "wt")
    print("\nLeyendo datos...")
    datos = entrada.readline()
    while datos:
        if datos.upper().find("IAN,")!=-1:
            armenia.write(datos.title()+"\n") #title pone el primer caracter en masyucula
        elif datos.upper().find("INI,")!=-1:
            italia.write(datos.title()+"\n")
        elif datos.upper().find("EZ,")!=-1:
            españa.write(datos.title()+"\n")
        datos = entrada.readline()
    print("Archivos generados correctamente")
except FileNotFoundError:
    print("No se encontró el archivo de entrada")
except OSError as error:
    print("ERROR:",str(error))
finally:
    try:
        entrada.close()
        armenia.close()
        italia.close()
        españa.close()
    except NameError:
        pass
```
